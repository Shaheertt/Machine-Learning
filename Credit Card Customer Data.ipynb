## <span style = "color:orange"> Visualization </span>

### Plot a scatterplot of Avg_credit_Limit vs Total_Credit_Cards with respect to Total_calls_made

```python
df = df.rename(columns={'Avg_Credit_Limit': 'Avg_credit_Limit'})
```

### Plot a pairplot

```python
sns.pairplot(df)
```

### Create K Means model (Let k = 4)

```python
from sklearn.cluster import KMeans
```

```python
model=KMeans(n_clusters=4)
```

### Fit predict for our Dataset

```python
s=model.fit_predict(df)
```

```python
s
```

### Assign the predicted clusters to a new column in our dataset

```python
df['clusters']=s
```

```python
df.head()
```

## Visualize the clusters

```python
df['clusters'].unique()
```

```python
df0=df[df['clusters']==0]
df1=df[df['clusters']==1]
df2=df[df['clusters']==2]
df3=df[df['clusters']==3]
```

```python
sns.scatterplot(data=df0,x='Avg_credit_Limit',y='Total_Credit_Cards',color='red')
sns.scatterplot(data=df1,x='Avg_credit_Limit',y='Total_Credit_Cards',color='green')
sns.scatterplot(data=df2,x='Avg_credit_Limit',y='Total_Credit_Cards',color='blue')
sns.scatterplot(data=df3,x='Avg_credit_Limit',y='Total_Credit_Cards',color='orange')
```
